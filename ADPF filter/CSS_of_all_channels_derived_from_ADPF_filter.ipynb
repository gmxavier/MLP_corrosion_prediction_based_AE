```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_1=pd.read_excel('channel_1.xlsx')
data_1_ad1=data_1[(data_1.Amplitude<54)&(45<=data_1.Amplitude)&(0<data_1.Duration)&(data_1.Duration<1000)]
data_1_ad2=data_1[(data_1.Amplitude<60)&(54<=data_1.Amplitude)&(100<data_1.Duration)&(data_1.Duration<1000)]
data_1_ad3=data_1[(data_1.Amplitude<65)&(60<=data_1.Amplitude)&(300<data_1.Duration)&(data_1.Duration<1000)]
data_1_ad4=data_1[(65<=data_1.Amplitude)&(500<data_1.Duration)&(data_1.Duration<1000)]
data_1_ad=data_1_ad1.append(data_1_ad2).append(data_1_ad3).append(data_1_ad4)
data_1_ad=data_1_ad[(140<=data_1.Peak_Frequency)&(data_1.Peak_Frequency<300)]
data_1_ad.index=range(len(data_1_ad))
data_1_ad.sort_values('Time',inplace=True)
css_1=np.cumsum(data_1_ad.Signal_Strength)
x_1=data_1_ad.Time
y=css_1
plt.figure(figsize=(10,6))
line=plt.plot(x_1,y,color='r',antialiased=False,linewidth=4)

plt.xlabel('Time (Min)',fontsize=16)
plt.ylabel('CSS',fontsize=16)



plt.grid()
plt.savefig('./CSS_ADPF_ch1.jpg',dpi=600)
plt.show()
```
